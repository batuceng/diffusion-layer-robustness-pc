```python
path = "../dist_results_maxvals"


attacks = os.listdir(path)

result_list = []

idx = 1
for atk in attacks:
    print(atk, "ATTACK", idx, "/", 6)
    attack_path = os.path.join(path, atk)
    idx += 1
    
    models = os.listdir(attack_path)
    for m in models:
        print(m)
        model_path = os.path.join(attack_path, m)
        
        results = os.listdir(model_path)
        for dist in results:
            json_file_name = os.listdir(os.path.join(model_path, dist))[0]
            json_file_path = os.path.join(model_path, dist, json_file_name)
            
            with open(json_file_path, "r") as f:
                json_file = json.load(f)
                result_list.append(json_file)
                f.close()

df = pd.DataFrame(result_list)
```
