```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
```

```python
df_maxvals = pd.read_csv("../dist_results_maxvals.csv")
df_pointdp = pd.read_csv("../dist_results_pointdp.csv")
df = pd.concat((df_maxvals, df_pointdp))
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
groups = df.groupby(["model", "t_list"])[["Defended_acc"]].mean()
maxvals = groups.apply(lambda x: x.sort_values(ascending=False))
type(maxvals)
```
