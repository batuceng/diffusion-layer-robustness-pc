```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# EXP 0
textfile = "scripts/gridlogs.txt"
with open(textfile) as file:
    file_contents = file.read()

tlist = []
acclist = []
avglist = []


for i, line in enumerate(file_contents.split('\n')):
    if "T list" in line:
        t_arr = line[10:-1].strip().split(',')
        t_arr = np.array(t_arr, dtype=int)
        tlist.append(t_arr)
    if "Test defense :: test acc:" in line:
        # t_arr = line[10:-1].strip().split(',')
        # t_arr = np.array(t_arr, dtype=int)
        test_acc_str, test_avg_str = line.split(',')
        acclist.append(float(test_acc_str.split(':')[-1].strip()))
        avglist.append(float(test_avg_str.split(':')[-1].strip()))
        
tlist = np.array(tlist).reshape(-1, 3)
acclist = np.array(acclist).reshape(-1, 1)
avglist = np.array(avglist).reshape(-1, 1)

```

```python
# Nprmalize
mini, maxi = 0.75, acclist.max()
colors = (acclist - mini) / (maxi - mini) 
colors = colors.clip(min=0)

# Plot3d
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(tlist[:,0], tlist[:,1], tlist[:,2], cmap='jet', c=colors)
plt.legend(colors)
ax.set_xlabel('L0', fontsize=12)
ax.set_ylabel('L1', fontsize=12)
ax.set_zlabel('L2', fontsize=12)
```

```python
# Top10
for i in np.argsort(colors, axis=0)[:10]:
    print(f"Acc:{float(acclist[i]):2f}, L0:{tlist[i][0,0]}, L1:{tlist[i][0,1]}, L2:{tlist[i][0,2]}")
```
