```python
import matplotlib.pyplot as plt
import sys
import json
sys.path.append("../")
import torch
from dataset.modelnet40 import LABELS
%matplotlib widget
```

```python
data = torch.load("../data_attacked/cw/pointnet/ATK_2024_01_22__01_17_18.pt")
size = len(data)
```

```python
with open("../data_attacked/cw/pointnet/ATK_2024_01_22__01_17_18.json", "r") as file:
    result_dict = json.load(file)
    print(result_dict)
```

```python
idx = 3

item = data[idx]


pc_clean = item["pointcloud"]
pc_attacked = item["attacked"]
label = LABELS[item["cate"].item()]

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
print(pc_attacked.shape)
ax.scatter3D(pc_clean[:,0], pc_clean[:,1], pc_clean[:,2], label="clean")
ax.scatter3D(pc_attacked[:,0], pc_attacked[:,1], pc_attacked[:,2], label="attacked")
plt.legend()
print(label)
plt.show()
```
